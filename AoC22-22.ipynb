```python
from copy import deepcopy
f = open("input22.txt", "r")
matrix = []
#dirs[0] is vector for walking to the right, next is turned 90Â° right (walking down)
dirs = [(1,0), (0,1), (-1,0), (0,-1)] 
dirSymbols = ">v<^"
stopMatrixInput = False
for line in f:
    if(len(line)) < 2:
        stopMatrixInput = True
    if stopMatrixInput == False:
        matrix.append(line.strip("\n"))
    if(len(line) > 2) and stopMatrixInput == True:
        path = line.strip("\n")
matrix2 = deepcopy(matrix)
h = len(matrix)
print(h)
print(len(path))
pos = (matrix[0].index("."),0)
facing = 0

numString = ""
pathList = []
for c in path:
    if c.isdigit():
        numString = numString + c
    else:
        if len(numString):
            pathList.append(int(numString))
        pathList.append(c)
        numString = ""
if len(numString):
    pathList.append(int(numString))

for l in matrix:
    print(l)
#print(pathList)

def isWall(pos, matrix, f):
    p = wrap(pos, matrix, f)
    if matrix[p[1]][p[0]] == "#":
        return True
    else:
        return False


def isWrap(pos, matrix):
    if pos[1] < 0 or pos[1] >= len(matrix):
        return True
    if pos[0] < 0 or pos[0] >= len(matrix[pos[1]]):
        return True
    if matrix[pos[1]][pos[0]] == " ":
        return True
    return False

def findfirstX(pos, matrix):
    line = matrix[pos[1]]
    for i in range(len(line)-1):
        if line[i] != " ":
            return (i, pos[1])

def findlastX(pos, matrix):
    line = matrix[pos[1]]
    for i in range(1,len(line)):
        if line[i] == " " and line[i-1] != " ":
            return (i-1, pos[1])
    return (len(line)-1, pos[1])

def findfirstY(pos, matrix):
    for i in range(len(matrix)-1):
        if matrix[i][pos[0]] != " ":
            return (pos[0],i)

def findlastY(pos, matrix):
    for i in range(1,len(matrix)):
        if len(matrix[i]) <= pos[0]:
            return (pos[0],i-1)
        if matrix[i][pos[0]] == " " and matrix[i-1][pos[0]] != " ":
            return (pos[0],i-1)
    return (pos[0], len(matrix)-1)
        

def wrap(pos, matrix, facing):
    if pos[1] < 0:
        return findlastY(pos, matrix)
    if pos[1] >= len(matrix):
        return findfirstY(pos, matrix)
    if pos[0] < 0:
        return findlastX(pos, matrix)
    if pos[0] >= len(matrix[pos[1]]):
        return findfirstX(pos, matrix)
    if matrix[pos[1]][pos[0]] == " ":
        if facing == 0:
            return findfirstX(pos, matrix)
        if facing == 1:
            return findfirstY(pos, matrix)
        if facing == 2:
            return findlastX(pos, matrix)
        if facing == 3:
            return findlastY(pos, matrix)
    return pos


for step in pathList:
    if type(step) == str:
        if step == "R":
            facing = (facing+1) % 4
        elif step == "L":
            facing = (facing-1) % 4
    elif type(step) == int:
        for i in range(step):
            newpos = (pos[0] + dirs[facing][0], pos[1] + dirs[facing][1])
            if isWall(newpos, matrix, facing):
                break
            else:
                matrix[pos[1]] = matrix[pos[1]][:pos[0]] + dirSymbols[facing] + matrix[pos[1]][pos[0]+1:] 
                pos = wrap(newpos, matrix, facing)
                


for l in matrix:
    print(l)

print("Position: X= " + str(pos[0]+1) + " Y= " + str(pos[1]+1))
print(1000 * (pos[1]+1) + 4 * (pos[0]+1) + facing)
```
